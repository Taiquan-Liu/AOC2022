```python
def exclusion_zone(sensor, beacon):
    d = abs(sensor[0] - beacon[0]) + abs(sensor[1] - beacon[1])
    exclude_coords = set()
    for dx in range(-d, d+1):
        dy_m = d - abs(dx)
        for dy in range(-dy_m, dy_m+1):
            exclude_coords.add((sensor[0]+dx, sensor[1]+dy))
    exclude_coords.remove(beacon)
    return exclude_coords
```
