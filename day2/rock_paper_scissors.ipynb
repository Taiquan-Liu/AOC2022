```python
mapping_op = {
    "A": "R",
    "B": "P",
    "C": "S",
}
mapping_me = {
    "X": "R",
    "Y": "P",
    "Z": "S",
}
rules = {
    "R": {
        "R": 3,
        "P": 0,
        "S": 6,
    },
    "P": {
        "R": 6,
        "P": 3,
        "S": 0,
    },
    "S": {
        "R": 0,
        "P": 6,
        "S": 3,
    },        
}
base_scores = {
    "R": 1,
    "P": 2,
    "S": 3,
}
```

```python
scores_me = [
    rules[pair[1]][pair[0]] + base_scores[pair[1]]
    for pair in matches
]
scores_me
```

```python
matches_new = [
    (
        mapping_op[pair.split(" ")[0]], 
        mapping_me_new[pair.split(" ")[1]](mapping_op[pair.split(" ")[0]])
    ) 
    for pair in s.split("\n")
]
matches_new
```

```python
scores_me_new = [
    rules[pair[1]][pair[0]] + base_scores[pair[1]]
    for pair in matches_new
]
scores_me_new
```
